```python
#importing library
import pandas as pd
```

# Create the dataframe
* A dataframe is a table used to display data

```python
data={
    'Name':['Alice','Susan','Bob','Mike','Jame','Peter','Kevin','Andrew','Martin','Adrian'],
    'Age':[12,7,8,17,15,18,14,8,9,10],
    'Score':[20,50,23,78,90,45,67,89,100,30]
}
#create a dataframe
df=pd.DataFrame(data)
#display the dataframe or print(df)
df
```

# Displaying a Few Columns in Data
* Display the first part of data using pandas.

```python
df.head()
```

#Check for Null Values
* Null Values are empty values in data.

```python
df.isnull().sum()
```

##Findings
*There are no null values in the dataframe
